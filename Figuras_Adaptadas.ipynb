```python
## MANIPULAÇÃO DOS DADOS
import pandas as pd
import numpy as np

## PLOTS
import matplotlib.pylab as plt
import seaborn as sns
from matplotlib import lines
%matplotlib inline
from matplotlib import ticker
from matplotlib.pylab import rcParams
from scipy.stats import norm
```

```python
## DADOS EM UM DF

dados = pd.DataFrame({'anos': ['2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', 
                               '2014', '2015', '2016', '2017', '2018', '2019', '2020'],
                        'capacidade': [22, 230, 241, 341, 601, 928, 1524, 2518, 3894, 5973, 8727, 
                                       10728, 12795, 14736, 15481, 17747]})
```

```python
rcParams['figure.figsize'] = 15, 6
plt.style.use('seaborn-whitegrid')

dados.sort_values('anos', ascending = True).plot(x = 'anos', y = 'capacidade', kind = 'bar', 
                                                    color = 'cyan', edgecolor = 'black')

plt.ylabel(''); plt.yticks(size = 14); plt.xlabel(''); plt.xticks(size = 14)
plt.title(' Evolução da Capacidade Instalada em MW ', size = 20);

#plt.savefig("CapacidadeInstalada.png", dpi = 300, bbox_inches='tight')
```
